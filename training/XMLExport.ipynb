```python
from os import listdir
import re
from time import time
```

```python
folder = "/Users/florian/Desktop/GTSDB/"

boxes = {}
our_sign = set()

def parse_i(s):
    return int(re.match(r"\d+", s).group())
        
with open(folder + 'gt2.txt') as f:
    for x in f:
        els = x.strip().split(";")
        i = parse_i(els[0])
        
        ns = map(int, els[1:])
        left, top, right, bottom, sign = ns
        rect = (left, top, right, bottom)
        
        if sign == 13:
            if i in our_sign:
                boxes[els[0]].append(rect)
            else:
                boxes[els[0]] = [rect]
                our_sign.add(i)
```

```python
print """<?xml version='1.0' encoding='ISO-8859-1'?>
<?xml-stylesheet type='text/xsl' href='image_metadata_stylesheet.xsl'?>
<dataset>
<name>imglab dataset</name>
<comment>Created by imglab tool.</comment>
<images>"""

for filename, rect in boxes.iteritems():
    print "  <image file='%s'>" % filename.replace(".ppm", ".jpg")
    for (left, top, right, bottom) in rect:
        print "    <box top='%s' left='%s' width='%s' height='%s'/>" % (top, left, right - left, bottom - top)
    print "  </image>"
    
print """</images>
</dataset>"""
```

```python
import cv2

new_folder = "/Users/florian/Desktop/GTSDB-dlib/"

for filename in boxes:
    img = cv2.imread(folder + filename)
    cv2.imwrite(new_folder + filename.replace(".ppm", ".jpg"), img)
```
